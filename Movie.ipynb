```python
import requests
import time
import json

API_KEY = "f295149057688867be603021005a275c"
# 1. GET THE DICTIONARY ONCE (English names)
genre_url = f"https://api.themoviedb.org{API_KEY}&language=en-US"
genre_data = requests.get('https://api.themoviedb.org/3/genre/movie/list').json()
genre_map = {g['id']: g['name'] for g in genre_data['genres']}

all_movies = []

# 2. THE 500-PAGE MOVIE LOOP
for page in range(1, 501):
    print(f"Fetching page {page}...")
    url = f"https://api.themoviedb.org{API_KEY}&page={page}"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.status_code.json()
        for movie in data['results']:
            # 3. THE INSTANT SWAP (happens in your RAM)
            movie['genre_names'] = [genre_map.get(gid) for gid in movie.get('genre_ids', [])]
            all_movies.append(movie)
    
    # 4. PROTECT YOUR KEY (The 0.05s safety pause)
    time.sleep(0.5)

# 5. SAVE YOUR DATASET
with open('tmdb_dataset.json', 'w') as f:
    json.dump(all_movies, f)
```
